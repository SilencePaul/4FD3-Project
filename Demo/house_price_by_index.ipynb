```python
import pandas as pd
df = pd.read_csv('data/housing_index.csv')
df.tail(30)
```

```python
df_filtered = df.drop_duplicates('COORDINATE')
locations = df_filtered.loc[df['New housing price indexes'] == 'House only']['GEO'].tolist() # Get unique cities
print(locations)
```

```python
def get_increase_rate(purchase_month, location):
  df_range = df[(df['GEO']==location) & (df['New housing price indexes']=='House only') & (df['REF_DATE'] >= purchase_month)]
  df_range = df_range.sort_values(by='REF_DATE', ascending=False)
  print(df_range['VALUE'].tolist())
  return df_range['VALUE'].iloc[0] / df_range['VALUE'].iloc[-1]

rate = get_increase_rate('1982-08', 'Hamilton, Ontario')
price = 300000
print(rate*price)
```
